# Slack Graph Visualization with yfiles-jupyter-graphs
Slackグラフ（NetworkX）を *yfiles-jupyter-graphs* で可視化するサンプルです。
- 既存のGraphMLがあれば読み込み、無ければDBからグラフを構築します。
- API仕様に合わせて GraphWidget(graph=G) + set_graph_layout('organic') を使用します。


```python
# ランタイムに必要なパッケージをインストール
%pip install -q yfiles-jupyter-graphs networkx
```

```python
import os, sys
import networkx as nx
sys.path.append('src')
from slack_graph.storage import Storage
from slack_graph.graph import build_graph

# notebooks/ から見た相対パス
graphml_path = '../output/graph.graphml'
if os.path.exists(graphml_path):
    G = nx.read_graphml(graphml_path)
else:
    store = Storage(os.getenv('DB_PATH', 'data/slack_graph.db'))
    G = build_graph(store)

print(f'Graph loaded: nodes={G.number_of_nodes()}, edges={G.number_of_edges()}')
```

```python
# yfiles-jupyter-graphs で表示
from yfiles_jupyter_graphs import GraphWidget
gw = GraphWidget(graph=G)
gw.set_graph_layout('organic')
gw
```
